```python
import hashlib

# P19. Genera el hash (SHA-256)
# Simulamos el contenido del cuaderno para este ejercicio de setup
contenido_cuaderno = b"Contenido binario del cuaderno para hash - Evidencia S1"
sha256 = hashlib.sha256(contenido_cuaderno).hexdigest()

print(f"SHA-256 generado: {sha256}")
```

```python
import json

# P20. Crea registro WORM (JSON)
registro = {
    "archivo": "00_setup_TuApellidoTuNombre.ipynb",
    "sha256": sha256,
    "fecha": datetime.datetime.now().isoformat()
}

nombre_json = "registro_hash.json"

# Guardamos el archivo en el entorno de Colab
with open(nombre_json, "w") as f:
    json.dump(registro, f, indent=4)

print(f"Archivo {nombre_json} creado con éxito.")
print(json.dumps(registro, indent=4))
```
